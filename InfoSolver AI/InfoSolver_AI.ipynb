# 🧠 InfoSolver AI
**Project Overview:**  
InfoSolver AI is a multi-tool intelligent agent that can answer questions using **Wikipedia**, **DuckDuckGo**, and **Calculator** tools. It also remembers past interactions, performs calculations, and returns clean Markdown answers.  
This notebook demonstrates the step-by-step building of this agent.

### 1️⃣ Installing Dependencies and Import Libraries
All necessary Python libraries and modules for LangChain, Groq LLM, tools, and memory are imported in this section.

```python
# =========================
# INSTALL DEPENDENCIES
# =========================
# pip install langchain langchain-community langchain-openai langchain-groq duckduckgo-search wikipedia numexpr
```

```python
# =========================
# IMPORTS
# =========================
import os
from dotenv import load_dotenv
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)

from langchain.agents import initialize_agent, Tool
from langchain.agents.agent_types import AgentType

from langchain_groq import ChatGroq

from langchain.tools import WikipediaQueryRun
from langchain.utilities import WikipediaAPIWrapper

from langchain_community.tools import DuckDuckGoSearchResults
from langchain.chains import LLMMathChain

from langchain.memory import ConversationBufferMemory
```

### 2️⃣ Initialize LLM (Groq)
Set up the Groq language model with the chosen model name (`openai/gpt-oss-120b`), temperature, and token limits.  
This is the core model the agent will use for reasoning and generating responses.

```python
# =========================
# SET API KEY (GROQ)
# =========================
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
# =========================
# INITIALIZE LLM (GROQ)
# =========================
llm = ChatGroq(
    model="openai/gpt-oss-120b",
    temperature=0
)
```

### 3️⃣ Initialize Tools
**Description:**  
Configure the three main tools:  
- **Wikipedia Tool:** For general knowledge  
- **DuckDuckGo Tool:** For recent or live information  
- **Calculator Tool:** For numeric operations  

```python
# =========================
# CREATE TOOLS
# =========================

# 📚 Wikipedia Tool
wiki = WikipediaAPIWrapper()
wiki_tool = WikipediaQueryRun(api_wrapper=wiki)

wikipedia_tool = Tool(
    name="Wikipedia",
    func=wiki_tool.run,
    description="Useful for general knowledge questions"
)

# 🌐 DuckDuckGo Search Tool
search = DuckDuckGoSearchResults()

duckduckgo_tool = Tool(
    name="DuckDuckGo Search",
    func=search.run,
    description="Useful for web search and current information"
)

# 🧮 Calculator Tool
math_chain = LLMMathChain.from_llm(llm=llm)

calculator_tool = Tool(
    name="Calculator",
    func=math_chain.run,
    description="Useful for mathematical calculations"
)
```

```python
# =========================
# COMBINE TOOLS
# =========================
tools = [
    wikipedia_tool,
    duckduckgo_tool,
    calculator_tool
]
```

```python
# =========================
# ADD MEMORY
# =========================
memory = ConversationBufferMemory(
    memory_key="chat_history",
    return_messages=True
)
```

### 4️⃣ Configure Agent
**Description:**  
Combine the LLM, tools, and memory into a LangChain agent.  
Here, we also configure parsing rules and set it to return clean Markdown answers.

```python
# =========================
# INITIALIZE AGENT
# =========================
agent_kwargs = {
    "system_message": "Always return only the final answer. Do NOT include thoughts, actions, or observations. Format the response cleanly in Markdown."
}

agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.CONVERSATIONAL_REACT_DESCRIPTION,
    memory=memory,
    verbose=False,
    handle_parsing_errors=True,
    agent_kwargs=agent_kwargs
)
```

### 5️⃣ Run Interactive Agent
**Description:**  
Provide a code block to take user input and run it through the agent.  
The agent will choose the right tool(s), perform calculations if needed, and return a final answer in Markdown.

```python
# =============
# RUN AGENT 
# =============

query = "Who is Imran Khan?"

response = agent.run(query)
print(f"Agent: {response}")
```
